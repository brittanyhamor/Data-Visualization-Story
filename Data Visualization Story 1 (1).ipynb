```r
library(lubridate)
library(dplyr)
library(ggplot2)
```

```r
cancer <- read.csv("Race Cancer Data .csv")
```

Cancer is the second leading cause of death in Nebraska and has increased three spots in the United States ranking from 2014 to 2015, according to the Center for Disease Control and Prevention.   

The graph shows the top 10 cancers with the highest mortality rate by each race. Rates are the average annual number of cases per 100,000 people, excluding gender-specific sites (female breast, prostate, ovary, cervix uteri, corpus uteri)

```r
library(repr)
options(repr.plot.width=15, repr.plot.height=10)
ggplot(cancer, aes(Cancer.Type, weight=Rate, fill=Race)) + geom_bar() + scale_color_brewer("Race") + scale_fill_discrete(name="Race") + 
labs(x="Cancer Types", y="Rate", title="Top 10 Cancer Mortality Rates by Race in Nebraska", subtitle="African-Americans have a higher mortality rate for most cancers in Nebraska", caption="Source: Nebraska Department of Health and Services | By Brittany Hamor") +
theme(plot.title = element_text(size = 26, face = "bold"), axis.title = element_text(size = 18, face = "bold"), legend.text = element_text(size = 16), legend.position = "top", legend.title = element_text(size = 16), axis.text = element_text(size = 14), plot.subtitle = element_text(size = 20), axis.text.x = element_text(angle = 90, hjust = 1))
```

```r
TopCancers <- read.csv("TopThree.csv")
```

The three cancers that significantly affect all races are breast, colorectal (colon), lung and bronchus. African-Americans have the highest mortality rate in all three cancers, even with lower incident rates.

```r
ggplot(TopCancers, aes(Cancer.Type, weight=Rate, fill=Race)) + geom_bar() + scale_color_brewer("Race") + scale_fill_discrete(name="Race") + 
labs(x="Cancer Types", y="Rate", title="Breast, Colorectal, Lung and Bronchus Affect each Race in Nebraska", subtitle="African-Americans have the highest mortality rate for all of the three cancers that affect each race in Nebraska", caption="Source: Nebraska Department of Health and Services | By Brittany Hamor") +
theme(plot.title = element_text(size = 26, face = "bold"), axis.title = element_text(size = 18, face = "bold"), legend.text = element_text(size = 16), legend.position = "top", legend.title = element_text(size = 16), axis.text = element_text(size = 14), plot.subtitle = element_text(size = 20))
```

For example, prostate cancer and lung/bronchus have the top two highest mortality rates for African-Americans. This cancer is also the second and third most deadly for Caucasians and are in the top ten for the remaining races. Although prostate and lung/bronchus cancer has the highest mortality rate among African-Americans, it has a significantly lower incident rate than for Caucasians. 

Lung and bronchus cancer had the highest death rates for all races in total from 2004 to 2013. This cancer affected 11,668 Caucasians, 441 African-Americans, 47 American Indians, 72 Asian/Pacific Islanders and 169 Hispanics.

During the years 2004-2013, this cancer killed 44.9 percent of Caucasians, 63.6 percent of African-Americans, 58.2 American Indians, 24.7 percent of Asian/Pacific Islanders and 19 percent of Hispanics with the disease. 

```r
LungCancer <- read.csv("Lung Cancer.csv")
```

```r
library(repr)
options(repr.plot.width=12, repr.plot.height=10)
ggplot(LungCancer, aes(x=Race, weight=Death.Rate)) + geom_bar(fill="red")+ coord_flip() +
labs(x="Race", y="Count", title="Lung and Bronchus Cancer Mortality Rates by Race", subtitle="African-Americans have the highest mortality rate, even though Whites have a higher incident rate.", caption="Source: Nebraska Department of Health and Services | By Brittany Hamor") +
theme(plot.title = element_text(size = 15, face = "bold"), axis.title = element_text(size = 14, face = "bold"), axis.text = element_text(size = 12), plot.subtitle = element_text(size = 12))
```

```r
library(repr)
options(repr.plot.width=12, repr.plot.height=10)
ggplot(LungCancer, aes(x=Race, weight=Incident)) + geom_bar(fill="blue")+ coord_flip() +
labs(x="Race", y="Count", title="Lung and Bronchus Cancer Incidents by Race", subtitle="Whites have the highest incident count of all races.", caption="Source: Nebraska Department of Health and Services | By Brittany Hamor") +
theme(plot.title = element_text(size = 15, face = "bold"), axis.title = element_text(size = 14, face = "bold"), axis.text = element_text(size = 12), plot.subtitle = element_text(size = 12))
```

This cancer affected more than double the amount of Caucasians compared to African-Americans, but the African-American mortality rate was nearly 19 percent higher than for Caucasians.

Also, between the years 2004 to 2013, 11,986 Caucasians were diagnosed with prostate cancer, while only 521 African-Americans were diagnosed with the same cancer. 34.7 percent of the incidents for African-Americans resulted in death, while Caucasians death totaled 22.8

Inequalities in cancer mortality rates can be seen with this data set, but The Nebraska Department of Human Services does not think this is associated with race, but rather a more complex issue. 

The causes of these inequalities are complex and are believed to reflect social and economic disparities rather than biological differences associated with race, according to a study done by The Nebraska Department of Human Services in April 2017, Socioeconomic disparities include inequalities in work, wealth, income, education, housing, and overall standard of living, as well as barriers to high-quality cancer detection, early detection, and treatment services. 

The Nebraska Department of Human Services believes these are the preventive measures needed to be taken to decrease the number of lives lost, especially for the African-American community.

As such, cancer prevention and control strategies for African-Americans will almost certainly require a multi-pronged approach, including the adoption of healthier lifestyles and better access to and utilization of healthcare services, and will need to be implemented in a way that respects the unique cultural characteristics of the African-American community, according to a study done by The Nebraska Department of Human Services in April 2017.
